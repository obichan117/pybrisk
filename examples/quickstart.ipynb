## OHLC Data

```python
ticker = pb.Ticker("7203")  # Toyota

# Daily candles
df = ticker.ohlc()
df.head()
```

```python
# 5-minute candles
df_5m = ticker.ohlc("5m")
df_5m.head()
```

## Margin Lending Data

```python
margin = ticker.jsfc(count=30)
margin.head()
```

## Market-Wide Data

```python
market = pb.Market()

# All TSE stocks
info = market.stocks_info()
print(f"Total stocks: {len(info)}")
info.head()
```

```python
# Curated stock lists
lists = market.stock_lists()
for name, codes in lists.items():
    print(f"{name}: {len(codes)} stocks")
```

```python
# Market condition alerts
alerts = market.alerts()
alerts.head()
```

```python
# Trading schedule
market.schedule()
```
